# Notebook 01 - Data Acquisition

Este notebook apenas contém a documentação da estrutura do dataset e código renomeando as colunas.

```python
import pandas as pd

df = pd.read_csv('../data/raw/train.csv')

df.head()
```

## Largando colunas que não possuem utilidade

Irei dropar colunas que acrescentam pouco ou não acrescentam nada ao projeto.

As colunas que serão dropadas são:
* customer_id
* name
* gender

```python
df = df.drop(columns=['customer_id', 'name', 'gender'])

df.head()
```

## Renomeando as colunas

```python
colunas_nomes = {
    "age": "idade",
    "owns_car": "tem_carro",
    "owns_house": "tem_casa",
    "no_of_children": "numero_de_filhos",
    "net_yearly_income": "renda_anual",
    "no_of_days_employed": "dias_no_emprego",
    "occupation_type": "tipo_de_profissão",
    "total_family_members": "total_membros_familia",
    "migrant_worker": "imigrante",
    "yearly_debt_payments": "pagamento_anual_de_debitos",
    "credit_limit": "limite",
    "credit_limit_used(%)":"porcentagem_uso_credito",
    "prev_defaults":"atrasos_anteriores",
    "default_in_last_6months": "atrasos_nos_ultimos_6meses",
    "credit_card_default": "inadimplencia"

}

df.rename(columns=colunas_nomes, inplace=True)

df.head()
```
